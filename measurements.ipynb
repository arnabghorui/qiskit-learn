```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import EstimatorV2 as Estimator
```

```python
# create quantum circuits with 3 qubits
qc = QuantumCircuit(3)

# apply Hadamard gate to qubit 2
qc.h(2)

# perform controlled-y gate between qubit 1 and qubit 2, with 2 being the control
qc.cy(2, 1)

# perform controlled-x gate between qubit 0 and qubit 1, with 1 being the control
qc.cx(0, 1)

# measure the qubit 1
qc.measure_all()


```

```python
qc.draw("mpl")
```

```python
# get the measuremet result, for one shot
qc.measure_all()
```
